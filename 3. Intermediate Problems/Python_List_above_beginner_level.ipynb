## 2Ô∏è‚É£ Find Second Largest Number

```python
nums = [10, 20, 4, 45, 98, 99, 5]
# Your code here
n=len(nums)
for i in range(n):
  if i==2:
    break
  for j in range(n-i-1):
    if nums[j]>nums[j+1]:
      temp=nums[j]
      nums[j]=nums[j+1]
      nums[j+1]=temp

print(nums[-2])
```

```python
list1 = [1, 3, 5, 7]
list2 = [2, 4, 6, 8]

newlist = []

n1 = len(list1)
n2 = len(list2)

m=0
n=0

while(m<n1 and n<n2):
  if list1[m] < list2[n]:
    newlist.append(list1[m])
    m+=1
  else:
    newlist.append(list2[n])
    n+=1

while(m < n1):
  newlist.append(list1[m])
  m+=1

while(n < n2):
  newlist.append(list2[n])
  n+=1

print(newlist)
```

## 9Ô∏è‚É£ Find Most Frequent Element

```python
nums = [1, 3, 1, 3, 2, 1, 4, 1, 5, 1]
# Your code here
d={}

for i in nums:
  d[i]=d.get(i,0)+1
print(d)

maximum=max(d, key=d.get)
print(maximum)
```
