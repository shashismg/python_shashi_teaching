### ❓ Longest Consecutive Sequence
**Difficulty:** Medium

Given an unsorted list of integers, write a function to return the length of the longest consecutive elements sequence.

Example:
Input: [100, 4, 200, 1, 3, 2]
Output: 4
# Explanation: The longest consecutive sequence is [1, 2, 3, 4]

---

```python
# Write your solution here
nums=[100, 4, 200, 1, 3, 2,5,300,400,500]
length = len(nums)
for i in range(length):
    for j in range(length-i-1):
      if nums[j]>nums[j+1]:
        temp=nums[j]
        nums[j]=nums[j+1]
        nums[j+1]=temp
print(nums)
count=0
n=nums[0]
m=nums[-1]

while n<m:
  if n in nums:
    n+=1
    count+=1
  else:
    break

print(count)
```

### ❓ Find the Duplicate Number
**Difficulty:** Medium

Given a list of integers where each number is in the range [1, n] inclusive and there are n + 1 integers, find the duplicate number.

Example:
Input: [3, 1, 3, 4, 2]
Output: 3

---

```python
# Write your solution here
num=[3, 1, 3, 4, 4, 4, 2]

s=set()
duplicat=set()
for i in num:
  if i in s:
    duplicat.add(i)
  else:
    s.add(i)
print(list(duplicat))
```

### ❓ Find All Pairs with Given Sum
**Difficulty:** Medium

Given a list of integers and a target sum k, return all unique pairs of numbers that sum up to k.

Example:
Input: nums = [1, 3, 2, 2, 4, 0, 5], k = 4
Output: [(1, 3), (2, 2), (0, 4)]

---

```python
nums = [1, 3, 2, 2, 4, 0, 5]
l=len(nums)
l1=[]
k=4

for i in range(l):

  for j in range(i+1,l):
    if nums[i]+nums[j]==k:

      l1.append((nums[i],nums[j]))

print(l1)


```

### ❓ Majority Element
**Difficulty:** Medium

Given a list of integers, find the element that appears more than ⌊n/2⌋ times. Assume such an element always exists.

Example:
Input: [2, 2, 1, 1, 1, 2, 2]
Output: 2

---

```python
# Write your solution here
nums=[2, 2, 1, 1, 1, 2, 2]
d={}

for i in nums:
  d[i]=d.get(i,0)+1
print(d)

maximum=max(d, key=d.get)
print(maximum)
```
