# 🧠 Python Basics & Logic Assignment

This assignment includes questions to test your basic Python coding skills and logical thinking. Answer all questions using Python.

### Q1: Find the maximum of three numbers

### Q2: Check if a number is even or odd

### Q10: Check if a number is an Armstrong number
