### Q1: Print a right-angled triangle of stars
```
*
* *
* * *
```


````python
n=3
for i in range(1,n+1):
  for j in range(i):
    print(' * ', end= "")
  print()
````

````python
n=5
for i in range(n):
  for j in range(n):
    print(" * ", end="")
  print()
````

````python
#bubble sort
nums=[3,6,1,9,8,22,56]
l=len(nums)

for i in range(l):
  for j in range(0,l-i-1):
    if nums[j]>nums[j+1]:
      temp=nums[j]
      nums[j]=nums[j+1]
      nums[j+1]=temp

print(nums)


````
