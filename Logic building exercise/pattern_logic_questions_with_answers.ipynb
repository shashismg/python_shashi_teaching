# ⭐ Pattern-Based Coding Questions

These are 10 pattern-based coding questions designed to test your logical thinking and Python skills. Try solving each before looking at the solution.

```python
# Q1: Print a right-angled triangle of stars
# *
# * *
# * * *
n = 5
for i in range(1, n + 1):
    print('* ' * i)
```

```python
# Q2: Print a number triangle
# 1
# 1 2
# 1 2 3
n = 5
for i in range(1, n + 1):
    for j in range(1, i + 1):
        print(j, end=' ')
    print()
```

```python
# Q3: Inverted star triangle
# * * *
# * *
# *
n = 5
for i in range(n, 0, -1):
    print('* ' * i)
```

```python
# Q4: Right-angled triangle with increasing numbers
# 1
# 2 3
# 4 5 6
n = 5
count = 1
for i in range(1, n + 1):
    for j in range(i):
        print(count, end=' ')
        count += 1
    print()
```

```python
# Q5: Triangle with alternating 1s and 0s
# 1
# 0 1
# 1 0 1
n = 5
for i in range(1, n + 1):
    for j in range(i):
        print((i + j) % 2, end=' ')
    print()
```

```python
# Q6: Floyd’s Triangle
# 1
# 2 3
# 4 5 6
n = 5
num = 1
for i in range(1, n + 1):
    for j in range(i):
        print(num, end=' ')
        num += 1
    print()
```

```python
# Q7: Pyramid with numbers
#     1
#    2 2
#   3 3 3
n = 5
for i in range(1, n + 1):
    print(' ' * (n - i) + (str(i) + ' ') * i)
```

```python
# Q8: Mirrored right triangle
#     *
#    * *
#   * * *
n = 5
for i in range(1, n + 1):
    print(' ' * (n - i) + '* ' * i)
```

```python
# Q9: Diamond shape pattern
#   *
#  * *
# * * *
#  * *
#   *
n = 3
for i in range(1, n + 1):
    print(' ' * (n - i) + '* ' * i)
for i in range(n - 1, 0, -1):
    print(' ' * (n - i) + '* ' * i)
```

```python
# Q10: Pascal’s Triangle (basic)
from math import comb
n = 5
for i in range(n):
    for j in range(n - i - 1):
        print(' ', end=' ')
    for j in range(i + 1):
        print(comb(i, j), end='   ')
    print()
```
