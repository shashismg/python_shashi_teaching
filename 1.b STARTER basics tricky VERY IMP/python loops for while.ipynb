```python
numbers = [10, 20, 30, 40, 50]
size = len(numbers)

# Using only index, not directly accessing list elements
for i in range(size):
    print(f"Index: {i}, Value: {numbers[i]}")
```

```python
numbers = [10, 20, 30, 40, 50]
size = len(numbers)

for number in numbers:
  print(number)

#odd position items
for odd_position in numbers[1::2]:
  print(odd_position)

#even position items
for even_position in numbers[::2]:
  print(even_position)
```
