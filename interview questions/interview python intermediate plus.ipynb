**Problem: Longest Consecutive Sequence**

Description:

Given an unsorted list of integers, write a function to return the length of the longest consecutive elements sequence.

The sequence should be made of consecutive numbers in value, not in position. The function should run in O(n) time if possible, but for your student, it's okay to first solve it with a simpler approach.

Example:
Input: [100, 4, 200, 1, 3, 2]
Output: 4

Explanation: The longest consecutive sequence is [1, 2, 3, 4]

Constraints:
You may assume all numbers are unique.
Input can be empty, in which case the result should be 0.
