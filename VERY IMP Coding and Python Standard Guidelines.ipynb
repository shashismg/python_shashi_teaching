**Coding standards**

* Use meaningful name for variables
* Make code readable and not complex
* Write step by step and dont write everything in one line
* One line one instruction
* Write comments where required





```python
a = ["apple, banana, grapes"] # give meaningful name like below

fruits = ["apple, banana, grapes"] # this will be easy to read

for i in a:
  print(i)

#instead write like below
for fruit in fruits:
  print(fruit)

```

```python
numbers = [10, 20, 30, 40, 50]

for i in range(len(numbers)):
    print(numbers[i])


#dont include too much of instruction in single line. break it line by line like below.
size = len(numbers)
for i in range(size):
    print(numbers[i])
```
